### Featuring engineering

```python
def preencher_nivel(gestor, nivel):
  if gestor == 1:
    return "Pessoa Gestora"

  else:
    return nivel
```

```python
dados['NOVO_NIVEL'] = dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```

```python
dados['NOVO_NIVEL'].value_counts()
```

```python
#pandas get_dummies - dados categóricos convertidos em dados numéricos

pd.get_dummies(dados, columns=['NIVEL'])
```

```python
def determinar_geracao(idade):
  if 39 < idade <= 58:
    return "Geração X"
  elif 29 < idade <= 39:
    return "Millenium"
  elif 13 < idade <= 29:
    return "Geração Z"
  else:
    return "Outra Geração"
```

```python
dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)
```

```python
dados['GERACAO'].value_counts()
```

```python
dados3['SALARIO'] = pd.to_numeric(dados3['SALARIO'].str.replace(',', '.'), errors='coerce')
```

```python
salario_por_idade = dados3.groupby('IDADE')['SALARIO'].mean()
salario_por_idade
```

```python
plt.figure()
plt.plot(salario_por_idade.index, salario_por_idade.values, marker = 'o', linestyle = '-')
plt.title('Salário Médio por Idade')
plt.xlabel('Idade')
plt.ylabel('Salário Médio')
plt.grid(True)
plt.show()
```

```python
#PLOTLY INTERATIVO

fig = px.line(salario_por_idade.reset_index(), x= 'IDADE', y= 'SALARIO', title='Salário Médio por Idade')
fig.show()
```

```python
plt.figure(figsize=(10,5))
plt.scatter(dados3['IDADE'], dados3['SALARIO'], alpha= 0.5)
plt.title('Relação entre Idade e Salário')
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.grid(True)
plt.show()
```
